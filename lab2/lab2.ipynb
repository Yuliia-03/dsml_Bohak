```python
import pandas as pd

data = pd.read_excel("./data/rollingsales_bronx.xlsx", skiprows=4)

data
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data[data['SALE PRICE'] == 0].groupby(["BOROUGH", "BLOCK"]).count()
```

```python
data = data.drop(data.loc[data['SALE PRICE'] == 0].index)
# data[data['SALE PRICE'] == 0].groupby(["BOROUGH", "BLOCK"]).count()
```

```python
data
```

```python
formated_data = data[(data['TOTAL UNITS'] > 0) & (data['LAND SQUARE FEET'] > 0)]
formated_data
```

```python
# formated_data.isnull().sum()
```

```python
formated_data = formated_data.dropna(axis='columns')
formated_data
# format_data.isnull().sum()
```

```python
formated_data["price per ft."] = formated_data["SALE PRICE"]/formated_data['LAND SQUARE FEET']
formated_data
```

```python
formated_data.groupby([formated_data["SALE DATE"].dt.year, formated_data["SALE DATE"].dt.month])[["SALE PRICE"]].sum().plot(kind="bar")
```

```python
formated_data.groupby([formated_data["SALE DATE"].dt.year, formated_data["SALE DATE"].dt.month])[["price per ft."]].mean().plot(kind="line")
```

```python
formated_data.groupby([formated_data["NEIGHBORHOOD"]])[["price per ft."]].mean().plot(kind="bar")
```
